```python
d1 = {1:'a',2:'b',3:'c'}
d1
```

```python
d1[1] #using keys can access values
```

```python
d1['a'] #by value can not access keys
```

```python
d1
```

```python
d1.get(1) #get() access value by keys pass as argument
```

```python
d2= d1.copy() #copy existing dict
```

```python
d2
```

```python
d2.items() #returns all items in dict
```

```python
d2.items('a') #can not return single item
```

```python
d2.fromkeys(d2, 2) #changes values by passing change as argument
```

```python
d2.fromkeys(1, 3) #cann't change single value
```

```python
d2.keys() #returns all keys in dict
```

```python
d2.keys(1) #not  takes argument
```

```python
d2.pop(2) #pop() removes keys value pair from dict by passing keys as an argument
```

```python
d2
```

```python
d3={1: 'a', 2: 'b', 3: 'c', 4:'d'}
d3
```

```python
d3.popitem() #popitem() remove and return key value pair no need of argument
```

```python
d3
```

```python
d3.popitem(1,'a')
```

```python
d3.values() #returns values in dict
```

```python
d3.values(1)
```

```python
d3.clear() #removes all elements in dict
d3
```
